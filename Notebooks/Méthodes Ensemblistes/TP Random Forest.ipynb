```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
rawDataTrain = pd.read_csv("../../DataSet/RandomForest/train.csv")
rawDataTest = pd.read_csv("../../DataSet/RandomForest/test.csv")
```

```python
rawDataTest.shape
```

```python
rawDataTrain
```

```python
y_train = rawDataTrain.values[:, -1]
y_test = rawDataTest.values[:, -1]
X_train = rawDataTrain.values[:, :-2]
X_test = rawDataTest.values[:, :-2]
```

```python
classifier = RandomForestClassifier(n_estimators=1000)
classifier.fit(X_train, y_train)
```

```python
#selection des features importantes 
from sklearn.feature_selection import SelectFromModel
model = SelectFromModel(classifier, prefit=True, threshold=0.008)
#model.fit(X_train, y_train)
train_x2 = model.transform(X_train)
test_x2 = model.transform(X_test)
train_x2.shape
#il reste 22 features sur les 563 initiales 
```

```python
from sklearn.metrics import accuracy_score
classifier2 = RandomForestClassifier(n_estimators=1000)
classifier2.fit(train_x2, y_train)
```

```python
y_pred_2 = classifier2.predict(test_x2)
outAccuracy2 = accuracy_score(y_test, y_pred_2)
print("Accuracy 2", outAccuracy2)
y_pred = classifier.predict(X_test)
outAccuracy = accuracy_score(y_test, y_pred)
print("Accuracy 1", outAccuracy)

```
